## инструкция как пользоваться кодом 

---
### Шаг 1:
1. На странице GitHub нажмите зелёную кнопку **`<> Code`** → **`Download ZIP`**.
2. Распакуйте архив в удобную папку.

 Если внутри архива есть ещё одна папка (например, `project-main/` или `source/`), зайдите в неё и ищите папку `game/`. Именно **эта** папка является корнем проекта.

---
### Шаг 2:
1. Скачайте актуальную версию: https://www.renpy.org/latest.html
2. Распакуйте архив в надёжное место (например, `C:\RenPy\` или `~/RenPy/`).
3. Запустите `renpy.exe` (Windows), `renpy.sh` (Linux) или `renpy.app` (macOS).
4. При первом запуске лаунчер попросит указать **папку для проектов**. Создайте отдельную директорию, например:
   - `C:\RenPyProjects\`
   - `~/Documents/RenPyGames/`

---
### Шаг 3:
Ren'Py находит проекты только в указанной при первом запуске папке.
1. Переместите **распакованную папку проекта** (ту, где сразу лежит `game/`) в вашу папку проектов.
2. Убедитесь, что путь **не содержит кириллицы и пробелов**.  
   ✅ `C:\RenPyProjects\MyVN\`  
   ❌ `C:\Пользователи\Имя\Новеллы\Моя Игра\` (может вызвать ошибки загрузки ресурсов)

---
### Шаг 4:
1. Перезапустите Ren'Py Launcher.
2. В списке проектов должна появиться ваша новелла.
3. Выберите её и нажмите **`Запустить`** (Launch / Run).
4. Лаунчер автоматически скачает нужные зависимости, если их нет, и откроет игру.

 Если проекта нет в списке: нажмите кнопку `Refresh` (Обновить) в лаунчере или проверьте, что папка `game/` находится в корне проекта, а не внутри другой папки.

Если всё совпадает → лаунчер увидит проект, и вы сможете запустить, тестировать или модифицировать его.
